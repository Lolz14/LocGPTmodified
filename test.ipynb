```python
from gnews import GNews
import json
```

```python
google_news = GNews(language='en', period='7d', max_results=100)
bank_news = google_news.get_news('Bank')
market_news = google_news.get_news("Market")
```

```python
all_news = bank_news + market_news
print(all_news)
```

```python
dataset = []
for news in all_news:
    current_news = google_news.get_full_article(
    news['url'])
    try:
        current_dict = {
            "title" : news["title"],
            "date"  : news["published date"],
            "publisher" : news["publisher"],
            "text"  : current_news.text
        }
    except:
        print(f'failed to scrap the article from {news["publisher"]}')
    else:
        dataset.append(current_dict)

with open('news_article.json', 'w') as outfile:
    json.dump(dataset, outfile)
```

```python
context = ""
for news in dataset:
    context += "\n+++++++++++++++++++++++++++++++++++\n" + news["text"]
print(context)
```
